```python
combined_financial_text = f"""
BALANCE SHEET DATA:
{balance_text}

CHART AND GRAPH DATA:
{chart_text}
"""

print(combined_financial_text[:1200])
```

```python
system_role = """
You are a professional financial analyst.
Your task is to analyze financial documents such as balance sheets,
income statements, and financial charts extracted from images.

You must identify key financial metrics, trends, risks, and outlooks,
and present them in a clear, investor-friendly summary.
"""
```

```python
output_guidelines = """
While summarizing, ensure the following structure:

1. Summary Title
2. Key Financial Metrics (Revenue, Profit, EPS, ROE, Ratios if available)
3. Income and Expense Analysis
4. Balance Sheet Highlights
5. Operational or Strategic Updates (if mentioned)
6. Risks, Challenges, and Market Conditions
7. Overall Financial Health and Outlook

Use simple language.
Use bullet points where appropriate.
Focus on decision-making insights.
"""
```

```python
final_prompt = f"""
{system_role}

Below is the extracted content from financial document images.

{combined_financial_text}

{output_guidelines}

Generate a concise yet detailed financial analysis summary.
"""
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
model_name = "google/flan-t5-large"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
max_input_length = 3000  # safe for flan-t5
trimmed_prompt = final_prompt[:max_input_length]

print("Prompt length:", len(trimmed_prompt))
```

```python
inputs = tokenizer(
    trimmed_prompt,
    return_tensors="pt",
    truncation=True
)
```

```python
outputs = model.generate(
    **inputs,
    max_new_tokens=400,
    do_sample=False
)

final_summary = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(final_summary)
```
